```python
class Articulo:
    nombre = ""
    cantidad = ""
    precio = ""

    def __init__(self, nombre, cantidad, precio):
        self.nombre = nombre
        self.cantidad = cantidad
        self.precio = precio
        
    def vender(self):
        total = self.cantidad * self.precio
        print(f"Nombre: {self.nombre} \nPrecio: ${self.precio} \nCantidad: ${self.cantidad} \nTotal: ${total}")

Doritos = Articulo("Doritos", 2, 1.25)
Doritos.vender()
```

```python
class Usuario:
    usuario = ""
    contrasena = ""
    correo = ""

    def __init__(self, usuario, contrasena, correo):
        self.usuario = usuario
        self.contrasena = contrasena
        self.correo = correo

    def inicarSesion(self, campoUsuario, campoContrasena):
        if (campoUsuario == self.usuario and campoContrasena == self.contrasena):
            print("Se inicio sesion correctamente!!!")
            return True
        elif(campoUsuario == self.correo and campoContrasena == self.contrasena):
            print("Se inicio sesion correctamente!!!")
            return True
        else:
            print("El usuario o la contraseña es incorrecta, por favor intentelo de nuevo")
            return False



usuario1 = Usuario("Daniel", "Univo2023", "dani_dr23@outlook.com")

while True:
    usuario = input("Ingrese el usuario o el correo para iniciar sesion: ")
    contrasena = input("Ingrese la contraseña: ")

    if usuario1.inicarSesion(usuario, contrasena):
        break
```

```python
class Estudiante:
    nombre = ""
    apellido = ""
    carnet = ""
    carrera = ""

    def __init__(self, nombre, apellido, carnet, carrera):
        self.nombre = nombre
        self.apellido = apellido
        self.carnet = carnet
        self.carrera = carrera
    
    def mostrarDatos(self):
        print(f"DATOS DEL ESTUDIANTE \nNombre: {self.nombre} \nApellido: {self.apellido} \nN° de carnet: {self.carnet} \nCarrera: {self.carrera}")

    def cambiarNombre(self, nuevoNombre):
        self.nombre = nuevoNombre
        print("Se cambio el nombre satisfactoriamente!!!")

    def cambiarApellido(self, nuevoApellido):
        self.apellido = nuevoApellido
        print("Se cambio el apellido satisfactoriamente!!!")

    def cambiarCarnet(self, nuevoCarnet):
        self.carnet = nuevoCarnet
        print("Se cambio el carnet satisfactoriamente!!!")

    def cambiarCarrera(self, nuevaCarrera):
        self.carrera = nuevaCarrera
        print("Se cambio la carrera satisfactoriamente!!!")

estudiante01 = Estudiante("Daniel", "Diaz", "U20220436", "Ingenieria en Desarrollo de Software")

while True:
    print("MENU DE OPCIONES\n1) Mostrar los datos \n2) Cambiar el nombre del usuario \n3) Cambiar el apellido del usuario \n4) Cambiar el carnet del usuario \n5) Cambiar la carrera del usuario \n6) Salir")
    opcion = input("Elija una opcion del 1-6: ")

    if opcion == "1":
        estudiante01.mostrarDatos()
    
    elif opcion == "2":
        newNombre = input("Ingrese un nuevo nombre: ")
        estudiante01.cambiarNombre(newNombre)

    elif opcion == "3":
        newApellido = input("Ingrese un nuevo apellido: ")
        estudiante01.cambiarApellido(newApellido)

    elif opcion == "4":
        newCarnet = input("Ingrese un nuevo carnet: ")
        estudiante01.cambiarCarnet(newCarnet)

    elif opcion == "5":
        newCarrera = input("Ingrese un nuevo carrera: ")
        estudiante01.cambiarCarrera(newCarrera)

    elif opcion == "6":
        break

    else:
        print("Seleccione una opcion valida")
```
